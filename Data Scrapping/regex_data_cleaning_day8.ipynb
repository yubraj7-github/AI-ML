***What is Regex?<br>***
Regex (regular expression) is just a pattern used to find or remove specific types of text.<br>
Think of it like a smart search tool that can remove:<br>
Numbers,<br>
Symbols,<br>
Extra spaces,<br>
Special characters,<br>
Unwanted words,<br>
Emails,<br>
URLs<br>

You tell regex what pattern to look for, and it cleans your text.<br>

***Why Regex Is Good for Cleaning?<br>***
->Because instead of removing each character manually,<br>
you write one small pattern and it cleans everything for you.<br>
Example:<br>
[0-9] â†’ removes all digits<br>
[a-zA-Z] â†’ finds all letters<br>
\s+ â†’ removes extra spaces<br>

```python
#Write a clean_price() function that uses regex to remove currency symbols (â‚¹, $, Â£) and convert the result into a float.

import re

def clean_price(price):
    # 1. Remove currency symbols â‚¹ $ Â£
    cleaned = re.sub(r"[â‚¹$Â£]", "", price)
    
    # 2. Remove spaces before/after
    cleaned = cleaned.strip()
    
    # 3. Convert to float
    return float(cleaned)
    
print(clean_price("$500"))     # 500.0
print(clean_price(" â‚¹700 "))   # 700.0
print(clean_price("Â£850"))     # 850.0
```

```python
#Write a clean_rating() function that extracts only the numeric rating from text like â€œRated 4 out of 5â€.

import re

def clean_rating(text):
    # Find the first number (integer or decimal) in the text
    match = re.search(r"\d+(\.\d+)?", text)
    
    if match:
        return float(match.group())
    else:
        return None  # if no number found
        
print(clean_rating("Rated 4 out of 5"))       # 4.0
print(clean_rating("Rating: 3.5 stars"))     # 3.5
print(clean_rating("No rating available"))   # None
```

```python
#Apply all cleaning functions to a list of items and store the cleaned output in a list called cleaned_data.

import re

def clean_text(text):
    text = re.sub(r"#", "", text)            # remove # but keep word
    text = re.sub(r"[^\w\s]", "", text)      # remove symbols & emojis
    text = re.sub(r"\s+", " ", text).strip() # remove extra spaces
    return text

def clean_price(price):
    cleaned = re.sub(r"[â‚¹$Â£,]", "", price).strip()
    return float(cleaned)

def clean_rating(rating):
    match = re.search(r"\d+(\.\d+)?", rating)
    return float(match.group()) if match else None

def clean_item_dict(item):
    return {
        "title": clean_text(item.get("title", "")),
        "price": clean_price(item.get("price", "0")),
        "rating": clean_rating(item.get("rating", ""))
    }
raw_items = [
    {"title": " Amazing Phone!!! ðŸ˜ƒ #tech ", "price": "â‚¹25,999", "rating": "Rated 4.5 out of 5"},
    {"title": "Laptop!!! ðŸ’» #work ", "price": "$75,000", "rating": "Rated 4.0 out of 5"},
    {"title": "Smart Watch!!! âŒš #fitness", "price": "Â£12,500", "rating": "Rated 3.8 out of 5"}
]    
cleaned_data = [clean_item_dict(item) for item in raw_items]
print(cleaned_data)
```

```python
#Create a validation step: check if cleaned price is numeric, rating is between 1â€“5, and text is non-empty. 
#If not valid, store the item in a list called rejected_items.

# Sample cleaned items (from previous step)
cleaned_data = [
    {'title': 'Amazing Phone tech', 'price': 25999.0, 'rating': 4.5},
    {'title': 'Laptop work', 'price': '75000', 'rating': 4.0},      # price as string â†’ invalid
    {'title': '', 'price': 12500.0, 'rating': 3.8},                # empty title â†’ invalid
    {'title': 'Tablet', 'price': 15000.0, 'rating': 5.5}           # rating > 5 â†’ invalid
]

# Lists to store valid and rejected items
valid_items = []
rejected_items = []

for item in cleaned_data:
    price_valid = isinstance(item.get('price'), (int, float))
    rating_valid = isinstance(item.get('rating'), (int, float)) and 1 <= item['rating'] <= 5
    title_valid = bool(item.get('title'))  # non-empty
    
    if price_valid and rating_valid and title_valid:
        valid_items.append(item)
    else:
        rejected_items.append(item)

print("Valid items:")
print(valid_items)
print("\nRejected items:")
print(rejected_items)
```
