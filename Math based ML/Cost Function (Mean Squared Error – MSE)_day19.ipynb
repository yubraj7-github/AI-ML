### 1 What is a cost function?

A **cost function** is a rule that tells us **how wrong our model is**.

It gives **one number** that shows **how bad the prediction is**.

* **Small number** â†’ good model
* **Big number** â†’ bad model


### Very simple idea (real-life)

Imagine you are **throwing a ball at a target ðŸŽ¯**.

* Cost function = **distance from the target**
* Closer to target â†’ **low cost**
* Far from target â†’ **high cost**




### 2 Why do we need a cost function in Machine Learning?

1. **To measure how wrong the model is**

   * The cost function tells us how far our predictions are from the actual values.
   * Example: If the actual value is 6 and predicted is 5, the cost function will show a small error, and if predicted is 1, it will show a bigger error.

2. **To guide the model to learn better**

   * During training, the model **tries to reduce the cost**.
   * Smaller cost â†’ better predictions.

3. **To decide the direction of improvement**

   * The cost function helps in adjusting model parameters in the **right direction** (e.g., increase or decrease weights).

4. **To compare different models**

   * Models with lower cost are **better**.

**Example (simple):**

* Actual value = 6, Predicted = 5
* Cost (MSE) = (6 âˆ’ 5)Â² = 1 â†’ small error
* Predicted = 1
* Cost (MSE) = (6 âˆ’ 1)Â² = 25 â†’ big error

**Conclusion:**

> Cost function is needed to **measure error**, **train the model**, and **improve predictions**.




### 3 Mean Squared Error (MSE)

**Definition:**  
Mean Squared Error (MSE) is a cost function that measures the average of the squares of the errors between predicted values and actual values.It tells us how far the predictions are from the true values, giving more weight to bigger errors.

**Formula:**  
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$
Where:  
- $y_i$ = actual value  
- $\hat{y}_i$ = predicted value  
- $n$ = total number of samples

**Example:**  
- Actual value = 6, Predicted = 5 â†’ Error = 1  
- Square of error = $1^2 = 1$  
- If total samples = 9, MSE = $1/9 \approx 0.11$

**Key Points:**  
- MSE is always non-negative.  
- Smaller MSE â†’ better model.  
- Larger MSE â†’ worse model, especially for big mistakes.

### 5  Why do we square the error in MSE?

1. **To make all errors positive**

   * Some errors can be negative (prediction < actual).
   * Squaring ensures all errors contribute positively to the total cost.

2. **To give more weight to larger errors**

   * Bigger mistakes are punished more because their square is much larger.
   * Example:

     * Error = 2 â†’ Squared = 4
     * Error = 5 â†’ Squared = 25
   * This makes the model focus on reducing **big errors**.

3. **To make the function smooth for optimization**

   * Squared errors are **differentiable**, which helps gradient-based methods (like gradient descent) to **update model parameters** efficiently.


**Easy way to remember:**

> â€œSquaring errors makes them positive and punishes big mistakes more.â€

**Easy way to remember:**

3. **Interpretation:**

### 8 Why is MSE commonly used in Linear Regression?

**Answer:**

1. **Shows how wrong predictions are** â€“ bigger mistakes get more attention because errors are squared.
2. **Easy to calculate and optimize** â€“ itâ€™s smooth, so the model can easily learn the best line using gradient descent.
3. **Gives a clear average error** â€“ tells us on average how far predictions are from actual values.

**Easy way to remember:**

> MSE is used because it helps the model learn better and focuses more on big mistakes.


```python
# 9 Write Python code to calculate MSE using a loop.
# Sample data
actual = [6, 4, 7, 5]       # Actual values
predicted = [5, 4, 8, 5]    # Predicted values

# Initialize sum of squared errors
squared_error_sum = 0

# Loop through each value
for a, p in zip(actual, predicted):
    error = a - p            # Calculate error
    squared_error = error ** 2
    squared_error_sum += squared_error

# Calculate MSE
n = len(actual)
mse = squared_error_sum / n

print("Mean Squared Error (MSE):", mse)
```

```python
# 10 Write a NumPy-based implementation of MSE.
import numpy as np

# Sample data
actual = np.array([6, 4, 7, 5])
predicted = np.array([5, 4, 8, 5])

# Calculate MSE
mse = np.mean((actual - predicted) ** 2)

print("Mean Squared Error (MSE):", mse)
```

### 13 How do you integrate MSE into Gradient Descent?


1. **MSE tells us how wrong the model is**

   * Mean Squared Error measures the average squared difference between predicted and actual values.

2. **Gradient Descent uses MSE to improve the model**

   * Gradient Descent is like climbing down a hill to reach the lowest point (minimum error).
   * MSE is the â€œheightâ€ of the hill. Higher MSE â†’ higher â€œpositionâ€, lower MSE â†’ closer to minimum.

3. **Steps in simple words:**

   * **Step 1:** Calculate MSE using current model parameters.
   * **Step 2:** Find the **gradient (slope)** of MSE with respect to each parameter.
   * **Step 3:** Update the parameters **in the opposite direction of the gradient** to reduce MSE.
   * **Step 4:** Repeat until MSE is as small as possible.

**Easy analogy:**

> Think of a ball rolling down a hill. MSE is the height of the hill. Gradient Descent is the ball moving to the lowest point (minimum error).



### 14 What is the role of MSE in parameter optimization?

1. **Measures how wrong the model is**

   * MSE tells us how far the predictions are from the actual values.

2. **Guides the model to improve**

   * In optimization (like Gradient Descent), we **adjust parameters** to **minimize MSE**.

3. **Determines the direction and size of change**

   * Gradients of MSE show **which way and how much** to change parameters to reduce errors.

4. **Helps find the best parameters**

   * By repeatedly reducing MSE, the model learns the parameters that give **the best predictions**.

**Easy way to remember:**

> MSE acts like a â€œguideâ€ that tells the model how to adjust its parameters to make better predictions.



### 15 How does MSE change after each training iteration?

**Answer:**

1. **Initially, MSE is high**

   * At the start, the modelâ€™s predictions are usually far from the actual values, so MSE is large.

2. **MSE decreases as training progresses**

   * During each iteration, parameters are updated (e.g., using Gradient Descent) to reduce errors.
   * As predictions get closer to actual values, the MSE becomes smaller.

3. **MSE eventually stabilizes**

   * After many iterations, MSE reaches a **minimum** or very small value.
   * This means the model has learned the best parameters.

**Easy analogy:**

> Think of MSE as the height of a hill. With each step (iteration), you roll down closer to the lowest point (minimum MSE).

**Quick summary:**

* Start â†’ High MSE
* Middle â†’ MSE decreasing
* End â†’ MSE small and stable


### 16 What happens if outliers exist in data?


1. **MSE increases a lot**

   * Since MSE squares the errors, **large errors from outliers get heavily penalized**, making the MSE much larger than it would be otherwise.

2. **Model may be biased**

   * The model may **adjust parameters too much** to fit the outliers, which can **hurt overall prediction accuracy** for normal data points.

3. **Predictions become less reliable**

   * Outliers can pull the regression line away from the majority of the data, leading to **wrong predictions** for most points.

**Easy way to remember:**

> Outliers act like â€œgiant errorsâ€ that make MSE big and can confuse the model.



### 17 Computing MSE for Multiple Features

**Theory:**  
- In multiple features (e.g., \(x_1, x_2, ..., x_m\)), a linear regression model predicts:

$$
\hat{y} = w_1 x_1 + w_2 x_2 + ... + w_m x_m + b
$$

- Error for each data point:

$$
\text{Error} = y_i - \hat{y}_i
$$

- Mean Squared Error (MSE) is the **average of squared errors**:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**Steps:**  
1. Compute predicted values \(\hat{y}_i\) using all features and model parameters.  
2. Compute the error for each data point: \(y_i - \hat{y}_i\).  
3. Square all errors.  
4. Compute the average of squared errors over all data points.  

**Easy way to remember:**  
> No matter how many features, MSE is just the average of squared differences between predictions and actual values.


### 18 Why is MSE Differentiable?

**Definition:**  
Mean Squared Error (MSE) is defined as:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**Reason it is differentiable:**  
- MSE is made of **squared terms**, which are **smooth and continuous** functions.  
- Smooth functions have a **derivative at every point**, so we can compute gradients with respect to model parameters (weights and bias).

**Importance in optimization:**  
- Differentiability allows us to compute **gradients**, which tell the model **how to adjust parameters** to reduce MSE using Gradient Descent.

**Easy way to remember:**  
> Squaring makes MSE smooth, so we can find slopes and optimize the model.


### 19 How do you plot MSE vs epochs?
* **Epoch** = one full pass over the training data
* After each epoch, we **calculate MSE**
* We **store MSE values**
* Then we **plot Epochs on X-axis** and **MSE on Y-axis**

This shows **how error decreases as training goes on**.



```python
import matplotlib.pyplot as plt

# Example MSE values after each epoch
epochs = [1, 2, 3, 4, 5]
mse_values = [10, 6, 3.5, 1.8, 0.9]

# Plot MSE vs Epochs
plt.plot(epochs, mse_values)
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("MSE vs Epochs")
plt.show()
```

### 22 How do you visualize error distribution?

### 23 How does MSE behave with noisy data?


1. **MSE increases**

   * Noise in data means actual values have random variations.
   * Predictions are less accurate â†’ errors are bigger â†’ MSE becomes larger.

2. **Harder for the model to learn patterns**

   * Noise makes it difficult to fit the true relationship between features and target.

3. **Fluctuating MSE during training**

   * With noisy data, MSE may **not decrease smoothly**; it can go up and down between epochs.

4. **May cause overfitting if noise is high**

   * Model may try to fit noise instead of the real pattern, leading to **low training MSE but high test MSE**.


### Easy way to remember

> More noise â†’ bigger errors â†’ higher MSE.




### 24 Why does MSE increase sharply with outliers?


1. **MSE squares the errors**

   * Squaring makes larger errors grow **much faster** than smaller errors.

2. **Outliers have very large errors**

   * If a data point is far from the true value, its squared error becomes **huge**.

3. **Big errors dominate the average**

   * Even a single outlier can **significantly increase MSE**, because it is weighted more heavily than normal points.



### Easy example:

| Actual | Predicted | Error | Squared Error |
| ------ | --------- | ----- | ------------- |
| 5      | 5         | 0     | 0             |
| 6      | 5         | 1     | 1             |
| 7      | 5         | 2     | 4             |
| 50     | 5         | 45    | 2025          |

* Without the outlier, MSE = (0+1+4)/3 = 1.67
* With outlier, MSE = (0+1+4+2025)/4 â‰ˆ 508  huge increase


### Easy way to remember

> Squared errors **punish big mistakes**, so outliers make MSE spike sharply.



### 25 How do you compare models using MSE?


1. **Compute MSE for each model**

   * Use the same dataset (training or test) to calculate MSE for all models.

2. **Compare the MSE values**

   * **Lower MSE â†’ better model**
   * **Higher MSE â†’ worse model**

3. **Use consistent data**

   * Always compare models on the **same dataset** to be fair.

4. **Check training and test MSE**

   * Training MSE â†’ how well the model fits training data
   * Test MSE â†’ how well the model generalizes to unseen data



### Easy way to remember

> The model with the **smallest MSE** is usually the best because it makes the **least prediction error**.
