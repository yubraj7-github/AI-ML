```python
# 1 Write a Python function to normalize a dataset using Min–Max scaling (no libraries except NumPy).

import numpy as np

def min_max_normalize(data):
    
   # Find minimum and maximum
    min_val = np.min(data)
    max_val = np.max(data)
    
    # Apply min-max formula
    normalized = (data - min_val) / (max_val - min_val)
    
    return normalized


dataset = np.array([5, 10, 15, 20])
normalized_dataset = min_max_normalize(dataset)
print("Original:", dataset)
print("Normalized:", normalized_dataset)
```

```python
# 2 Write a function euclidean_distance(x1, x2) that returns the Euclidean distance between two data points.

import math

def euclidean_distance(x1, x2):
    distance = 0
    for i in range(len(x1)):
        distance += (x1[i] - x2[i]) ** 2
    return math.sqrt(distance)


p1 = [2, 3]
p2 = [5, 7]

print(euclidean_distance(p1, p2))
```

```python
# 3 Randomly initialize K centroids from your dataset and print their values

import random

def initialize_centroids(data, K):
    # randomly select K unique points from data
    centroids = random.sample(data, K)
    return centroids

# Example dataset (2D points)
dataset = [
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8]
]

K = 2

centroids = initialize_centroids(dataset, K)
print("Initial Centroids:")
for c in centroids:
    print(c)

```

```python
# 4 Assign each data point to the nearest centroid using your distance function and return the cluster labels

import math

def euclidean_distance(x1, x2):
    distance = 0
    for i in range(len(x1)):
        distance += (x1[i] - x2[i]) ** 2
    return math.sqrt(distance)


def assign_clusters(data, centroids):
    labels = []

    for point in data:
        distances = []

        for centroid in centroids:
            d = euclidean_distance(point, centroid)
            distances.append(d)

        # index of nearest centroid
        label = distances.index(min(distances))
        labels.append(label)

    return labels



data = [
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8]
]

centroids = [
    [2, 3],
    [3, 4]
]

labels = assign_clusters(data, centroids)
print("Cluster Labels:", labels)
```

```python
# 5 Modify your code so it supports any value of K dynamically (K ≥ 2).

import math

def euclidean_distance(x1, x2):
    distance = 0
    for i in range(len(x1)):
        distance += (x1[i] - x2[i]) ** 2
    return math.sqrt(distance)

def assign_clusters(data, centroids):
    labels = []

    for point in data:
        distances = []

        for centroid in centroids:
            d = euclidean_distance(point, centroid)
            distances.append(d)

        # index of nearest centroid
        label = distances.index(min(distances))
        labels.append(label)

    return labels


# Example dataset
data = [
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8],
    [8, 9]
]

# Try any K ≥ 2
centroids = [
    [1, 2],
    [6, 7],
    [8, 9]
]  # K = 3

labels = assign_clusters(data, centroids)
print("Cluster Labels:", labels)
```

```python
# 6 Plot the dataset and show centroid positions before clustering starts (without sklearn).

import matplotlib.pyplot as plt

# Example dataset (2D points)
data = [
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8],
    [8, 9]
]

# Initial centroids (before clustering, K = 3)
centroids = [
    [2, 3],
    [6, 7],
    [8, 9]
]

# Separate x and y coordinates for plotting
x_data = [point[0] for point in data]
y_data = [point[1] for point in data]

x_cent = [c[0] for c in centroids]
y_cent = [c[1] for c in centroids]

# Plot dataset points
plt.scatter(x_data, y_data, color='blue', label='Data Points')

# Plot centroids
plt.scatter(x_cent, y_cent, color='red',  label='Centroids')

# Add titles and labels
plt.title("Dataset and Initial Centroids (Before Clustering)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
