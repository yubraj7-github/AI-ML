# Covariance and Correlation 

### 1. What is covariance and what does it measure in a dataset?
- **Covariance** measures how two variables **change together** in a dataset.  
- If both variables tend to increase together, covariance is **positive**.  
- If one increases while the other decreases, covariance is **negative**.  
- **Limitation:** Its value depends on the units of the variables, so it’s hard to interpret the strength of the relationship.

---

### 2. What is correlation and how is it different from covariance?
- **Correlation** measures both the **strength** and **direction** of a linear relationship between two variables.  
- Unlike covariance, correlation is **standardized**, so it always lies between **-1 and +1**.  

**Key Differences:**

| Feature              | Covariance                              | Correlation                               |
|----------------------|----------------------------------------|------------------------------------------|
| Meaning              | Measures how variables vary together   | Measures strength & direction of relationship |
| Value Range          | Any value (depends on units)           | Always between -1 and +1                  |
| Interpretation       | Hard to interpret magnitude            | Easy to interpret magnitude               |
| Formula              | `cov(X,Y) = Σ((X-mean(X))*(Y-mean(Y))) / (n-1)` | `corr(X,Y) = cov(X,Y) / (std(X)*std(Y))` |

---

### 3. What is the range of correlation values and what do they indicate?
- Correlation values are always between **-1 and +1**.  
  - **+1** → Perfect positive relationship  
  - **0** → No linear relationship  
  - **-1** → Perfect negative relationship  

**Interpretation:**

| Correlation (r) | Meaning in simple words                  |
|-----------------|----------------------------------------|
| 0.8 to 1.0      | Strong positive relationship           |
| 0.5 to 0.8      | Moderate positive relationship         |
| 0.3 to 0.5      | Weak positive relationship             |
| 0               | No linear relationship                 |
| -0.3 to -0.5    | Weak negative relationship             |
| -0.5 to -0.8    | Moderate negative relationship         |
| -0.8 to -1.0    | Strong negative relationship           |

---

### 4. What does a positive covariance mean?
- A **positive covariance** means that **both variables tend to increase or decrease together**.  
- Example: Height and weight – generally, taller people weigh more.  

---

### 5. What does a negative correlation represent in real-world data?
- A **negative correlation** means that **when one variable increases, the other tends to decrease**.  
- Example: The number of hours spent watching TV and exam scores – usually, more TV hours → lower scores.

---

### 6. Why is covariance not normalized?
- Covariance is **not normalized** because it depends on the units of the variables.  
- This makes it hard to compare across datasets with different scales.  
- Correlation fixes this by **scaling covariance** with standard deviations.

---

### 7. Can correlation exist without causation? Explain with an example.
- **Yes**, correlation does **not imply causation**.  
- Two variables can move together **by coincidence or due to a third factor**, without one causing the other.  

**Example:**  
- Ice cream sales and drowning incidents are positively correlated (both increase in summer), but buying ice cream **does not cause drowning**.  
- The third factor (summer/temperature) affects both.

---

### 8. What does a correlation value close to 0 indicate?
- A correlation close to **0** means **no linear relationship** between the two variables.  
- Changes in one variable **do not predict** changes in the other.  
- Example: Shoe size and exam scores – they are basically unrelated.
