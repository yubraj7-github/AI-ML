```python
""""Task 1: Calculate Basic Statistics from a List
• Write code to calculate:
– Mean
– Median
– Mode
– Minimum & Maximum
– Range
• Use your own list of 15 random numbers.""""
```

```python
#mean
random_number=[12,5,9,18,7,14,21,3,11,6,19,8,15,2,10]
mean_value=sum(random_number)/len(random_number)
print("Mean:", mean_value)
```

```python
#median
sorted_number=sorted(random_number)
print("sorted_number:",sorted_number)
n=len(sorted_number)
middle=n//2
if n % 2 == 0:
    median_value = (sorted_number[middle- 1] + sorted_number[middle]) / 2
else:
    median_value = sorted_number[middle]
print("Median:", median_value)
```

```python
#mode
frequency = {}   

for num in random_number:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

mode_value = max(frequency, key=frequency.get)
print("Mode:", mode_value)
```

```python
#minimum andmaximum
minimum_value =random_number[0]
maximum_value =random_number[0]

for num in random_number:
    if num < minimum_value:
        minimum_value = num
    if num > maximum_value:
        maximum_value = num
print("Minimum:", minimum_value)
print("Maximum:", maximum_value)
```

```python
#range
range_value = maximum_value - minimum_value
print("Range:", range_value)
```

```python
"""Task 2: Create a Frequency Distribution
• Write code that counts how many times each value appears in a list.
• Print frequency in ascending order.
• Use a list of 30 numbers."""

numbers = [12, 45, 23, 12, 45, 67, 23, 78, 12, 45,
           89, 23, 45, 56, 78, 12, 90, 23, 45, 67,
           12, 34, 78, 23, 45, 12, 67, 78, 23, 45]

frequency = {}   
for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1


print("Frequency Distribution:\n")

for value in sorted(frequency):
    print(value, ":", frequency[value])
```

```python
""""Task 3: Generate Random Data & Analyse
• Write code to generate 50 random integers between 1 and 100.
• From the generated data, calculate:
– Mean
– Variance
– Standard deviation""""

import random

# Generate 50 random integers between 1 and 100
data = [random.randint(1, 100) for i in range(50)]
print("Random Data:", data)

# Mean
mean = sum(data) / len(data)
print("Mean:", mean)

# Variance
variance = sum((x - mean) ** 2 for x in data) / len(data)
print("Variance:", variance)

#Standard Deviation
std_dev = variance ** 0.5
print("Standard Deviation:", std_dev)
```

```python
"""Task 4: Visualize Data
• Write code to plot a histogram of randomly generated numbers.
• Write code to plot a line graph of cumulative frequency.
• Use any data of your choice."""

import random
import matplotlib.pyplot as plt

data = [random.randint(1, 100) for _ in range(50)]
plt.figure(figsize=(10,5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.show()

sorted_data = sorted(data)

# cumulative frequency
cumulative_freq = []
cum_sum = 0
for num in sorted_data:
    cum_sum += 1 
    cumulative_freq.append(cum_sum)

plt.figure(figsize=(10,5))
plt.plot(sorted_data, cumulative_freq, marker='o', color='green')
plt.title("Cumulative Frequency Graph")
plt.xlabel("Value")
plt.ylabel("Cumulative Frequency")
plt.grid(True)
plt.show()
```

```python
"""Task 5: Compare Two Datasets
• Create two different lists of numbers (size 20 each).
• Write code to:
– Compare means of both lists
– Compare standard deviations
– Find which dataset is more “spread out"""

import random
data1=[random.randint(1,100) for i in range(20)]
data2=[random.randint(1,100) for i in range(20)]
print('Data1:',data1)
print('Data2:',data2)

#mean
mean1=sum(data1)/len(data1)
mean2=sum(data2)/len(data2)
print("Mean of Dataset 1:", mean1)
print("Mean of Dataset 2:", mean2)

#compare mean
if mean1>mean2:
      print("Dataset 1 has a higher mean.")
elif mean1 < mean2: 
    print("Dataset 2 has a higher mean.")
else:
    print("Both datasets have the same mean.")

#standard deviation
#we have to calculate the variance first
variance1 = sum((x - mean1)**2 for x in data1) / len(data1)
variance2 = sum((x - mean2)**2 for x in data2) / len(data2)

#standard deviation
std_dev1 = variance1 ** 0.5
std_dev2 = variance2 ** 0.5

print("Standard Deviation of Dataset 1:", std_dev1)
print("Standard Deviation of Dataset 2:", std_dev2)

#compare spread
if std_dev1 > std_dev2:
    print("Dataset 1 is more spread out.")
elif std_dev1 < std_dev2:
    print("Dataset 2 is more spread out.")
else:
    print("Both datasets have the same spread.")
```
