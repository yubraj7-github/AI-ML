```python
'''Task 1 — Load CSV (No Pandas) + Missing Value Report
Load a CSV file using only open() and csv.reader.
Then generate a report showing how many missing/empty values are present in each column.'''

import csv

#  Open CSV file
filename ="C:/Users/shres/OneDrive/Documents/train.csv"
with open(filename, 'r', newline='') as csvfile:
    #Converts the file into a CSV reader object
    reader = csv.reader(csvfile)
    
    # Read header
    header = next(reader)
    
    #Initialize missing value counters for each column
    missing_counts = [0] * len(header)
    
    #Loop through each row
    for row in reader:
        for i, value in enumerate(row):
            # Consider value missing if empty or only spaces
            if value.strip() == "":
                missing_counts[i] += 1

# Print missing value report
print("Missing Value Report:")
for c, ct in zip(header, missing_counts):
    print(f"{c}: {ct}")
```

```python
'''Task 3 — Find Mode (Most Frequent Value) Manually
Using only dictionaries and loops, compute the mode of a chosen numeric column.
If multiple modes exist, print all of them.'''

import csv

filename ="C:/Users/shres/OneDrive/Documents/train.csv"
numeric_column = "Age"

#  Read CSV
values = []
with open(filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    col_index = header.index(numeric_column)
    
    for row in reader:
        value = row[col_index].strip()
        if value != "":           
            try:
                values.append(float(value))
            except:
                pass             

# Count frequency using dictionary
freq = {}
for v in values:
    if v in freq:
        freq[v] += 1
    else:
        freq[v] = 1
# Find maximum frequency
max_freq = max(freq.values())
# Find all modes
modes = []
for key in freq:
    if freq[key] == max_freq:
        modes.append(key)

# Print result
print(f"Mode(s) of column '{numeric_column}': {modes}")
```

```python
'''Task 4 — Implement Custom Quartile Function
Write a function get_quartile(data, q) that returns any quartile (Q1, Q2, Q3).
Use it to compute and print Q1, median (Q2), and Q3.'''

def get_quartile(data, q):
  
    sorted_data = sorted(data)
    n = len(sorted_data)
    
    # Compute the position
    pos = q * (n + 1) / 4
    
    #Split position into integer and fractional part
    k = int(pos)          # integer part
    d = pos - k           # fractional part
    
    # Handle edge cases
    if k == 0:
        return sorted_data[0]
    elif k >= n:
        return sorted_data[-1]
    else:
        # Step 4: Linear interpolation
        return sorted_data[k-1] + d * (sorted_data[k] - sorted_data[k-1])


data = [10, 12, 15, 18, 25, 30]

Q1 = get_quartile(data, 1)
Q2 = get_quartile(data, 2)  # median
Q3 = get_quartile(data, 3)

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
```

```python
'''Task 5 — Outlier Detection Using IQR Rule
Using your quartile results, compute:
• IQR = Q3 − Q1
• lower_bound = Q1 − 1.5 × IQR
• upper_bound = Q3 + 1.5 × IQR
Detect and print all outliers.'''

def get_quartile(data, q):
    sorted_data = sorted(data)
    n = len(sorted_data)
    pos = q * (n + 1) / 4
    k = int(pos)
    d = pos - k
    if k == 0:
        return sorted_data[0]
    elif k >= n:
        return sorted_data[-1]
    else:
        return sorted_data[k-1] + d * (sorted_data[k] - sorted_data[k-1])


data = [10, 12, 15, 18, 25, 30]

# Compute quartiles
Q1 = get_quartile(data, 1)
Q2 = get_quartile(data, 2)  # median
Q3 = get_quartile(data, 3)

# Compute IQR and bounds
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = []
for value in data:
    if value < lower_bound or value > upper_bound:
        outliers.append(value)

# Print results
print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
'''Task 6 — Manual Skewness Calculation
Write a function to calculate the skewness of a numeric column using the formula:
skewness = (Σ((x − mean)³) / n) / (std_dev³)
Using only loops and basic math operations.
After calculating, state if the distribution is positively or negatively skewed.'''

def calculate_skewness(data):
    n = len(data)
    
    # Step 1: Calculate mean
    mean = sum(data) / n
    
    # Step 2: Calculate standard deviation
    variance = sum((x - mean)**2 for x in data) / n
    std_dev = variance ** 0.5
    
    # Step 3: Calculate numerator: sum((x - mean)^3)/n
    cubed_dev_sum = sum((x - mean)**3 for x in data)
    numerator = cubed_dev_sum / n
    
    # Step 4: Skewness
    skewness = numerator / (std_dev**3)
    
    return skewness


data = [10, 12, 15, 18, 25, 30]

skew = calculate_skewness(data)
print("Skewness:", skew)

# Step 5: Interpret
if skew > 0:
    print("Distribution is positively skewed")
elif skew < 0:
    print("Distribution is negatively skewed")
else:
    print("Distribution is roughly symmetric")
```

```python
'''Task 7 — Manual Histogram Binning (Text-Based Chart)
Create 6–10 bins for a numeric column.
Count how many values fall inside each bin manually.
Print a text-based histogram such as:
0–10: *****
10–20: ***
(Not allowed to use matplotlib or numpy.)'''

data = [5, 7, 12, 15, 18, 22, 25, 27, 30, 33, 35, 40]  # example numeric data

# decide number of bins
num_bins = 6

# find min & max
minimum = min(data)
maximum = max(data)

# find bin size
bin_size = (maximum - minimum) / num_bins

# create empty bins (list of counters)
bins = [0] * num_bins

# count values manually
for value in data:
    index = int((value - minimum) / bin_size)
    
    # Handle edge case: last value goes out of range
    if index == num_bins:
        index -= 1
    
    bins[index] += 1

# Print histogram
start = minimum
for i in range(num_bins):
    end = start + bin_size
    print(f"{round(start)}–{round(end)}: " + "*" * bins[i])
    start = end
```

```python
'''Task 8 — Manual Correlation Matrix (3 Columns)
Select any three numeric columns.
Manually compute pairwise correlation between each pair.
Print the result as a small matrix format.'''
# Example numeric dataset
A = [10, 20, 30, 40, 50]
B = [15, 25, 35, 45, 55]
C = [50, 40, 30, 20, 10]

columns = {"A": A, "B": B, "C": C}
names = list(columns.keys())

# Function to calculate correlation between two lists
def correlation(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    
    numerator = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    denominator = (sum((x[i] - mean_x)**2 for i in range(n))**0.5) * \
                  (sum((y[i] - mean_y)**2 for i in range(n))**0.5)
    
    return numerator / denominator

# Compute correlation matrix
matrix = []
for i in range(3):
    row = []
    for j in range(3):
        if i == j:
            row.append(1.0)  # correlation with itself
        else:
            row.append(round(correlation(columns[names[i]], columns[names[j]]), 2))
    matrix.append(row)

# Print matrix
print("\t" + "\t".join(names))
for i, row in enumerate(matrix):
    print(f"{names[i]}\t" + "\t".join(map(str, row)))
```

```python
'''Task 9 — Manual Linear Regression (x → y)
Choose any two numeric columns (x and y).
Manually compute:
• variance(x)
• covariance(x, y)
• slope (m)
• intercept (b)
• predicted y values
• MSE of predictions
No numpy allowed.'''
#  numeric columns
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

n = len(x)

# mean of x and y
mean_x = sum(x) / n
mean_y = sum(y) / n

# Step 2: variance of x
variance_x = sum((xi - mean_x)**2 for xi in x) / n

# Step 3: covariance of x and y
cov_xy = sum((x[i] - mean_x)*(y[i] - mean_y) for i in range(n)) / n

# Step 4: slope and intercept
m = cov_xy / variance_x
b = mean_y - m * mean_x

# Step 5: predicted y values
y_pred = [m*xi + b for xi in x]

# Step 6: Mean Squared Error
mse = sum((y[i] - y_pred[i])**2 for i in range(n)) / n

# Print results
print("Mean of x:", mean_x)
print("Mean of y:", mean_y)
print("Variance of x:", variance_x)
print("Covariance of x and y:", cov_xy)
print("Slope (m):", m)
print("Intercept (b):", b)
print("Predicted y:", y_pred)
print("Mean Squared Error (MSE):", mse)
```

```python
# Example numeric dataset
data_columns = {
    "A": [10, 12, 15, 18, 25, 30],
    "B": [5, 8, 12, 15, 20, 22]
}

# Custom quartile function
def get_quartile(data, q):
    sorted_data = sorted(data)
    n = len(sorted_data)
    pos = q * (n + 1) / 4
    k = int(pos)
    d = pos - k
    if k == 0:
        return sorted_data[0]
    elif k >= n:
        return sorted_data[-1]
    else:
        return sorted_data[k-1] + d * (sorted_data[k] - sorted_data[k-1])

# Function to compute mean
def mean(data):
    return sum(data)/len(data)

# Function to compute standard deviation
def std_dev(data):
    m = mean(data)
    return (sum((x - m)**2 for x in data)/len(data))**0.5

# Function to compute median using get_quartile
def median(data):
    return get_quartile(data, 2)

# Loop through each column
for col_name, col_data in data_columns.items():
    n = len(col_data)
    min_val = min(col_data)
    max_val = max(col_data)
    mean_val = mean(col_data)
    q1 = get_quartile(col_data, 0.25)
    q3 = get_quartile(col_data, 0.75)
    med = median(col_data)
    std = std_dev(col_data)
    
    # Summary
    print(f"\nColumn: {col_name}")
    print(f"Count: {n}, Min: {min_val}, Max: {max_val}")
    print(f"Mean: {mean_val:.2f}, Median: {med}, Std Dev: {std:.2f}")
    print(f"Q1: {q1}, Q3: {q3}")
    
    # Text-based visualization
    vis = []
    for val in col_data:
        if val < q1:
            vis.append('L')
        elif val <= q3:
            vis.append('M')
        else:
            vis.append('H')
    print("Distribution:", ' '.join(vis))
```
