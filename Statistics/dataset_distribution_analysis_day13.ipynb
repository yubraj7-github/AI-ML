***Descriptive Statistics***<br>
Descriptive statistics means summarizing data so that we can easily understand it.<br>
Instead of looking at many numbers, we describe them using a few values.<br>

Examples:<br>
Mean → average value<br>
Median → middle value<br>
Mode → most frequent value<br>
Range → max − min
Standard deviation → how spread out the data is<br>

It answers: “What does the data look like?”<br>

***Distribution Types***<br>
A distribution shows how data values are spread.<br>

1. Symmetric Distribution<br>
-Left side and right side look almost the same<br>
-Mean = Median<br>
-Data is balanced<br>
Example: 2, 4, 6, 8, 10<br>

2. Right-Skewed Distribution (Positive Skew)<br>
-Tail is longer on the right side<br>
-Mean > Median<br>
-Few large values pull the mean to the right<br>
Example: 2, 4, 6, 8, 100<br>

3. Left-Skewed Distribution (Negative Skew)<br>
-Tail is longer on the left side<br>
-Mean < Median<br>
-Few small values pull the mean to the left<br>
Example: 1, 5, 10, 20, 30

```python
#Implement the following functions without using NumPy or Pandas built-in statistical functions:
#• mean(data)
#• median(data)
#• mode(data)
#• variance(data)
#• std(data)

def mean(data):
    return sum(data) / len(data)

def median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]
def mode(data):
    freq = {}

    for x in data:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1

  
    return max(freq, key=freq.get)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)

def std(data):
    return variance(data) ** 0.5
```

```python
#Apply these functions to a numerical column from the Titanic dataset (for example: Age or Fare).
import csv

ages = []

filename ="C:/Users/shres/OneDrive/Documents/titanic.csv"
with open(filename, 'r',newline='') as f:
     reader = csv.DictReader(f)
     for row in reader:
        if row["Age"] != "":
            ages.append(float(row["Age"]))
                      
```

```python
#Print the following descriptive statistics:
#• total number of records
#• minimum and maximum values
#• mean, median, mode
#• variance and standard deviation
print("Total number of records:", len(ages))
print("Minimum Age:", min(ages))
print("Maximum Age:", max(ages))
print("Mean Age:", mean(ages))
print("Median Age:", median(ages))
print("Mode Age:", mode(ages))
print("Variance of Age:", variance(ages))
print("Standard Deviation of Age:", std(ages))
```

```python
#Based on the values, identify the distribution type of the selected column:
#• symmetric
#• left-skewed
#• right-skewed

median_val=median(ages)
mean_val=mean(ages)
if mean_val==median_val:
    print("symmetric")
elif mean_val>median_val:
    print("right-skewed")
else:
    print("left-skewed")
```

```python
#Compare mean and median values and explain (in comments or print output) what this comparison indicates about the data distribution.

if mean_val > median_val:
    print("Mean > Median  Right-skewed (positively skewed)")
elif mean_val < median_val:
    print("Mean < Median  Left-skewed (negatively skewed)")
else:
    print("Mean ≈ Median  Symmetric distribution")
```

```python
#After removing outliers (using your Day-1 outlier function), recompute mean and standard deviation and compare the results with the original values.
def remove_outliers(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    q1 = sorted_data[n // 4]
    q3 = sorted_data[(3 * n) // 4]
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    # Keep only values within bounds
    filtered_data = [x for x in data if lower <= x <= upper]
    return filtered_data
ages_no_outliers = remove_outliers(ages)

mean_original = mean(ages)
std_original = std(ages)

mean_no_outliers = mean(ages_no_outliers)
std_no_outliers = std(ages_no_outliers)

print("Original Mean:", mean_original)
print("Mean after removing outliers:", mean_no_outliers)
print("Original Standard Deviation:", std_original)
print("Std after removing outliers:", std_no_outliers)
```

```python
#Divide the data into bins (class intervals) and create a simple frequency table using plain Python (no Pandas groupby).

#Decide number of bins
num_bins = 5  
min_age = min(ages)
max_age = max(ages)
bin_width = (max_age - min_age) / num_bins
print("Bin width:", bin_width)

#Create bins
bins = []
for i in range(num_bins):
    lower = min_age + i * bin_width
    upper = lower + bin_width
    bins.append((round(lower, 1), round(upper, 1)))  # store as (lower, upper)

print("Bins:", bins)
#Count frequency in each bin
freq = []

for b in bins:
    count = sum(1 for x in ages if b[0] <= x < b[1])
    freq.append(count)

# Include the max value in the last bin
freq[-1] += sum(1 for x in ages if x == max_age)

print("Bin\tFrequency")
for i in range(num_bins):
    print(f"{bins[i][0]} - {bins[i][1]}\t{freq[i]}")
```
